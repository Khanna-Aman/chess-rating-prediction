```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from scipy.optimize import fsolve
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# ===============================
# 1️⃣ Fetch Rating Data from Lichess API
# ===============================

# Replace with your Lichess username
username = "Khanna-Aman"

# API endpoint for rating history
url = f"https://lichess.org/api/user/{username}/rating-history"

# Send GET request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    
    # Extract rating history for the desired format (e.g., Blitz, Bullet, Rapid, etc.)
    # Currently using Bullet rating history
    rating_format = next(item for item in data if item["name"] == "Bullet")
    
    # Extract dates and ratings
    dates = [datetime(year=entry[0], month=entry[1] + 1, day=entry[2]) for entry in rating_format["points"]]
    ratings = [entry[3] for entry in rating_format["points"]]
    
    # Save data to CSV dynamically based on the selected rating format
    filename = f"lichess_{rating_format['name'].lower()}_ratings.csv"
    with open(filename, "w") as f:
        f.write("Date,Rating\n")
        for date, rating in zip(dates, ratings):
            f.write(f"{date.strftime('%Y-%m-%d')},{rating}\n")
    
    print(f"✅ Rating history saved to {filename}")
else:
    print(f"❌ Failed to retrieve data: {response.status_code}")
    exit()

# ===============================
# 2️⃣ Load & Preprocess Data
# ===============================

# Load Data dynamically based on the selected rating format
df = pd.read_csv(filename)
df["Date"] = pd.to_datetime(df["Date"], format="mixed", dayfirst=True)  # Ensure proper date format
df = df.sort_values(by="Date")  # Sort data chronologically

# Convert date to numerical format (days since first recorded game)
df["Days"] = (df["Date"] - df["Date"].min()).dt.days

# Split Data into Training & Testing Sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(df["Days"].values.reshape(-1, 1), df["Rating"], test_size=0.2, shuffle=False)

# ===============================
# 3️⃣ Determine Best Polynomial Degree
# ===============================

best_degree = None
lowest_mse = float("inf")
best_model = None
best_poly = None

# Try polynomial degrees from 1 to 5
for degree in range(1, 6):
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Predict and calculate error
    y_pred = model.predict(X_test_poly)
    mse = mean_squared_error(y_test, y_pred)

    # Select the best degree with the lowest error
    if mse < lowest_mse:
        lowest_mse = mse
        best_degree = degree
        best_model = model
        best_poly = poly

print(f"\n✅ Best Polynomial Degree: {best_degree} (MSE: {lowest_mse:.2f})")

# ===============================
# 4️⃣ Train Model with Best Degree
# ===============================

poly = best_poly
model = best_model

# ===============================
# 5️⃣ Define Prediction Functions
# ===============================

def predict_rating(target_date):
    """
    Predicts the blitz rating for a given future date.
    User should enter the target date in 'dd-mm-yyyy' format.
    """
    target_days = (pd.to_datetime(target_date, dayfirst=True) - df["Date"].min()).days
    
    if target_days <= df["Days"].max():
        return "⚠️ Date must be in the future."
    
    X_future = poly.transform(np.array([[target_days]]))
    prediction = model.predict(X_future)[0]
    return round(prediction, 2)

def when_will_i_reach(target_rating):
    """
    Estimates the date when the user will reach a specified blitz rating.
    """
    if target_rating < df["Rating"].min():
        return "✅ You have already surpassed this rating!"

    # Extend forecast for 5 years
    future_days = np.linspace(df["Days"].min(), df["Days"].max() + 365*5, 500)
    X_future = poly.transform(future_days.reshape(-1, 1))
    future_ratings = model.predict(X_future)

    if max(future_ratings) < target_rating:
        return "⚠️ Based on the current trend, reaching this rating may not be possible. However, with consistent improvement and practice, you can definitely achieve it!"


    # Solve for the date where we reach target_rating
    def equation(days):
        X_future = poly.transform(np.array(days).reshape(-1, 1))
        return model.predict(X_future)[0] - target_rating

    estimated_days = fsolve(equation, df["Days"].max() + 50)[0]

    if not np.isfinite(estimated_days) or estimated_days < df["Days"].max():
        return "⚠️ Based on the current trend, reaching this rating may not be possible."

    estimated_date = df["Date"].min() + pd.Timedelta(days=int(estimated_days))
    return estimated_date.strftime("%d-%m-%Y")

# ===============================
# 6️⃣ Predict Future Ratings & Plot Graph
# ===============================

# Generate predictions for future dates
future_days = np.linspace(df["Days"].min(), df["Days"].max() + 365, 100)
X_future = poly.transform(future_days.reshape(-1, 1))
predicted_ratings = model.predict(X_future)

# Plot actual vs predicted ratings
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Rating"], label="Actual Ratings", color="blue")
plt.plot(df["Date"].min() + pd.to_timedelta(future_days, unit='D'), predicted_ratings, 
         label=f"Polynomial Regression (Degree {best_degree})", linestyle="dashed", color="green")
plt.xlabel("Date")
plt.ylabel("Rating")
plt.legend()
plt.title(f"{rating_format['name']} Chess Rating Projection")
plt.xticks(rotation=45)
plt.grid()
plt.show()

# ===============================
# 7️⃣ Example Usage (Uncomment & Modify)
# ===============================
# ⚠️ Run this in a separate code block to avoid re-training the model

# Predict rating on a specific future date
target_date = "01-06-2026"  # dd-mm-yyyy format
print(f"📅 Estimated rating on {target_date}: {predict_rating(target_date)}")

# Estimate when a certain rating will be reached
target_rating = 2750
print(f"🏆 Estimated date to reach {target_rating} rating: {when_will_i_reach(target_rating)}")
```
